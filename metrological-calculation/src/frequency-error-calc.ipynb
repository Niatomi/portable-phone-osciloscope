## Расчёт погрешности по частоте


### Расчёт ОУ

Так как частота сигнала варьируется от 0 до 200 МГц также используется плоскость для представления данных

Напряжение питания: 5 В

Slew Rate: 20 V/us


### Основная аддитивная


```python
Vs = 5
slew_rate = 20 * (10 ** (-6))
# slew_rate = 0.000002
f = np.linspace(32768, 200_000_000, num=100, dtype=np.float64)
fig, ax = plt.subplots()


def inaccuracy_percent(slew_rate: float, freq: float, vs: float):
    return (1 - (slew_rate / (freq * vs))) * 100


x = f
y = np.abs(inaccuracy_percent(slew_rate, f, Vs))
ax.plot(x, y)
# ax.set_ylim(bottom=0.)
ax.set(xlabel='Frequency', ylabel='Error Percent',
       title='')
ax.ticklabel_format(useOffset=False)
ax.grid()

# fig.savefig("test.png")
plt.show()
print(inaccuracy_percent(slew_rate, 32768, 5))
```

### Расчёт ADC


```python
conversion_rate = 200 * 10 ** 6
conversion_time = conversion_rate
f = np.linspace(32768, 200_000_000, num=100, dtype=np.float64)
fig, ax = plt.subplots()


def inaccuracy_percent(freq: float, ct: float):
    return (1 - (1 / (freq * ct))) * 100


x = f
y = np.abs(inaccuracy_percent(f, conversion_time))
ax.plot(x, y)
# ax.set_ylim(bottom=0.)
ax.set(xlabel='Frequency', ylabel='Error Percent',
       title='')
ax.ticklabel_format(useOffset=False)
ax.grid()

# fig.savefig("test.png")
plt.show()
```
